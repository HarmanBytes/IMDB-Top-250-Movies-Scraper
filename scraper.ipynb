# Function

```python
def setup(driver_path):
    try:
        cService = webdriver.EdgeService(executable_path=path)
        driver = webdriver.Edge(service=cService)
        return driver
    except:
        print('something went wrong')
```

```python
def scrape_imdb_top_movies(driver_path):
    website = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
    driver = setup(driver_path)
    driver.get(website)
    row_dict = {
        'rank': np.nan,
        'title': np.nan,
        'release_year': np.nan,
        'duration': np.nan,
        'rate_type': np.nan,
        'star_rating': np.nan,
        'rated_by': np.nan
    }

    data = []

    try:
        containers = driver.find_elements(By.XPATH, value='//div[@class="ipc-metadata-list-summary-item__tc"]')
        try:
            for container in containers:
                row = row_dict.copy()
                try:
                    row['rank'], row['title'] = container.find_element(By.CLASS_NAME, value='ipc-title__text').text.split('.')
                except:
                    pass

                try:
                    movie_metadata_container = container.find_element(By.XPATH, value='.//div[@class="sc-300a8231-6 dBUjvq cli-title-metadata"]')
                    try:
                        metadata = movie_metadata_container.find_elements(By.XPATH, value='./span[@class="sc-300a8231-7 eaXxft cli-title-metadata-item"]')
                        row['release_year'] = metadata[0].text
                        row['duration'] = metadata[1].text
                        row['rate_type'] = metadata[2].text
                    except:
                        pass
                except:
                    pass

                try:
                    row['star_rating'] = container.find_element(By.XPATH, value='.//span[@class="ipc-rating-star--rating"]').text
                except:
                    pass
                try:
                    row['rated_by'] = container.find_element(By.XPATH, value='.//span[@class="ipc-rating-star--voteCount"]').text
                except:
                    pass  
                data.append(row)
                
        except:
            pass
    except:
        driver.quit()

    driver.quit()
    return data
            
```

```python
def list_to_csv(data):
    df = pd.DataFrame(data)
    df.to_csv('top250_movies_data.csv', index=False)
```

```python
data = scrape_imdb_top_movies("D:/Software/edgeDriver/msedgedriver.exe")
```
